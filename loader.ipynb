### Connecting to DATABASE

```python

import mysql.connector



# Connect to your MySQL database
conn = mysql.connector.connect(user='sqluser', password='password', host='localhost', database='BTC_sentiment')

# Create a cursor object
cursor = conn.cursor()


```

### Inserting TweetData to database

```python
from datetime import datetime


def ensure_date_dim(date_str):
    if date_str is None:

        cursor.execute("INSERT INTO DateDim (Date_recorded, Time_recorded) VALUES (NULL, NULL)")
        conn.commit()
        return cursor.lastrowid
    else:
        # Proceed as before if date_str is not None
        parsed_datetime = datetime.strptime(date_str, '%Y-%m-%dT%H:%M:%S.%fZ')
        date_part = parsed_datetime.date()  # Extracts the date part
        time_part = parsed_datetime.time()  # Extracts the time part

        # Insert the date and time into DateDim and return its DateID
        cursor.execute("INSERT INTO DateDim (Date_recorded, Time_recorded) VALUES (%s, %s)", (date_part, time_part))
        conn.commit()
        return cursor.lastrowid


def ensure_author_dim(author_twitter_id, author_name):
    cursor.execute("SELECT AuthorID FROM AuthorDim WHERE AuthorTwitterID = %s", (author_twitter_id,))
    author_id = cursor.fetchone()
    if author_id:
        return author_id[0]
    else:
        cursor.execute("INSERT INTO AuthorDim (AuthorName, AuthorTwitterID) VALUES (%s, %s)", (author_name, author_twitter_id))
        conn.commit()
        return cursor.lastrowid

def insert_tweet_dim(tweet_id, tweet_content):
        tweet_content = tweet_content[:49] if tweet_content is not None else ""
        cursor.execute(
            "INSERT INTO TweetDim (TweetID, TweetContent) VALUES (%s, %s) ON DUPLICATE KEY UPDATE TweetID = TweetID", 
            (tweet_id, tweet_content)
        )
        conn.commit()
```

```python

# Adjust the range and filename pattern as necessary
for day in range(2, 31):
    filename = f'tweets_with_sentiment_{day}.csv'
    df = pd.read_csv(filename, engine='python')
    
    for index, row in df.iterrows():
        tweet_id = row['Tweet ID']
        
        date_id = ensure_date_dim(row['Date Created'])
        author_id = ensure_author_dim(row['Author ID'], row['Author Username'])
        insert_tweet_dim(row['Tweet ID'], row['Text'])
        
        # Assuming PriceStart, PriceEnd, etc., are to be filled later or have placeholder values
        analysis_fact_query = """INSERT INTO AnalysisFact (TweetID, DateID, AuthorID, ReTweetCount, ReplyCount, LikeCount, QuoteCount, ImpressionCount, SentimentScore, WeightedSentiment) 
                                 VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s)"""
        analysis_fact_values = (row['Tweet ID'], date_id, author_id, row['Retweet Count'], row['Reply Count'], row['Like Count'], row['Quote Count'], row['Impression Count'], row['Sentiment Score'], 0) # Assuming WeightedSentiment is 0 for now
        try:
            cursor.execute(analysis_fact_query, analysis_fact_values)
            conn.commit()
        except mysql.connector.Error as e:
          print(f"Error inserting analysis fact for TweetID {tweet_id}: {e}")
```

### Inserting Price data to database

```python
from datetime import datetime,timedelta

def find_financial_record(tweet_datetime):
    end_range = tweet_datetime + timedelta(hours=1)
    filtered_df = financial_df.loc[tweet_datetime:end_range]
    if not filtered_df.empty:
        return filtered_df.iloc[0]  
    else:
        return None

# Example usage
print(find_financial_record(datetime.strptime('2023-01-30T09:00:01.000Z', '%Y-%m-%dT%H:%M:%S.%fZ')))
```

```python
for day in range(2, 30):
    filename = f'tweets_with_sentiment_{day}.csv'
    df = pd.read_csv(filename, engine='python')
    for index, row in df.iterrows():
        # Check if 'Date Created' is None before attempting to parse
        if pd.notna(row['Date Created']):
            tweet_datetime = datetime.strptime(row['Date Created'], '%Y-%m-%dT%H:%M:%S.%fZ')
            financial_record = find_financial_record(tweet_datetime)

            if financial_record is not None:
                # Convert each float64 to native float and NaN to None
                open_price = float(financial_record['OPEN']) if pd.notna(financial_record['OPEN']) else None
                close_price = float(financial_record['CLOSE']) if pd.notna(financial_record['CLOSE']) else None
                high_price = float(financial_record['HIGH']) if pd.notna(financial_record['HIGH']) else None
                low_price = float(financial_record['LOW']) if pd.notna(financial_record['LOW']) else None
                volume = float(financial_record['VOLUME']) if pd.notna(financial_record['VOLUME']) else None

                update_query = """
                UPDATE AnalysisFact
                SET PriceStart=%s, PriceEnd=%s, PriceHigh=%s, PriceLow=%s, Volume=%s
                WHERE TweetID=%s
                """
                
                try:
                    cursor.execute(update_query, (
                        open_price,
                        close_price,
                        high_price,
                        low_price,
                        volume,
                        row['Tweet ID']
                    ))
                    conn.commit()
                except mysql.connector.Error as e:
                    print(f"Error inserting analysis fact for TweetID {row['Tweet ID']}: {e}")
        else:
            # Handle rows with 'None' in 'Date Created', if necessary
            print(f"Skipping row due to missing 'Date Created' for TweetID {row['Tweet ID']}")
```
