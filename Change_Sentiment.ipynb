```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Read the CSV file
df = pd.read_csv('BTC-USD(1).csv')

# Convert 'Date' column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Filter data for the desired date range (2nd to 9th January)
start_date = pd.Timestamp('2023-01-02')
end_date = pd.Timestamp('2023-01-30')
df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]

# Plotting adjustments
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Adj Close', color=color)
ax1.plot(df['Date'], df['Adj Close'], color=color, marker='o', label='Adj Close')
ax1.tick_params(axis='y', labelcolor=color)

# Format the x-axis to show only the day
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%d'))

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Volume', color=color)
ax2.plot(df['Date'], df['Volume'], color=color, marker='x', label='Volume')
ax2.tick_params(axis='y', labelcolor=color)

# Adding title and legend
plt.title('Adj Close and Volume')
fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

sentiment = []
WeightedSentiment = []
ChangeInWS = []
ChangeInS = []

for x in range(2, 29):
    df2 = pd.read_csv(f'tweets_with_sentiment_{x}.csv', engine='python')
        # Convert 'Sentiment Score' and 'Like Count' to numeric, coerce errors
    df2['Sentiment Score'] = pd.to_numeric(df2['Sentiment Score'], errors='coerce')
    df2['Like Count'] = pd.to_numeric(df2['Like Count'], errors='coerce')
    
    # Optionally, fill NaN values with 0 or some other default value
    df2.fillna({'Sentiment Score': 0, 'Like Count': 0}, inplace=True)

    total_sentiment_score2 = df2['Sentiment Score'].sum()
    print(f"Sum of all sentiment scores on jan {x}:", total_sentiment_score2)

    df2['Weighted Sentiment'] = df2['Sentiment Score'] * df2['Like Count']
    WeightedSentiment2 = df2['Weighted Sentiment'].sum()
    print(f"Total Weighted Sentiment Score on jan {x}:", WeightedSentiment2)

    if sentiment and WeightedSentiment:
        changews = ((WeightedSentiment[-1] - WeightedSentiment2) / WeightedSentiment2) * 100
        changeS = ((sentiment[-1] - total_sentiment_score2) / total_sentiment_score2) * 100
    else:
        changews = 0
        changeS = 0

    ChangeInS.append(changeS)
    ChangeInWS.append(changews)
    WeightedSentiment.append(WeightedSentiment2)
    sentiment.append(total_sentiment_score2)

# Now your lists are properly initialized and populated, you can plot them

plt.plot(range(2, 29), ChangeInWS, marker='o', label='Change in Weighted Sentiment')
plt.plot(range(2, 29), ChangeInS, marker='x', label='Change in Sentiment')

plt.xlabel('Day of January')
plt.ylabel('Percentage Change')
plt.title('Change in Sentiment and Weighted Sentiment')
plt.legend()
plt.show()
```
